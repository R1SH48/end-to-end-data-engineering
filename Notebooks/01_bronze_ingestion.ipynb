```python
%sql
USE CATALOG dbacademy;
USE SCHEMA dbuser_100426;
```

```python
# Set your AWS credentials
access_key = ""
secret_key = ""

# set paths for files
s3_relative_path = f"s3://cloud-storage-s3-for-external-dw/folder-100426/archive-kaggle"
object_menu_items_csv = "/menu_items.csv"
object_restaurant_csv = "/restaurants.csv"
object_menu_items_parquet = "/menu_items.parquet"
object_restaurant_parquet = "/restaurants.parquet"
object_swiggy_india_jsonl = "/swiggy_india.jsonl"
```

```python
#reading the file into a dataframe
raw_menu_items_csv_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("fs.s3a.access.key", access_key) \
    .option("fs.s3a.secret.key", secret_key) \
    .load(s3_relative_path+object_menu_items_csv)

raw_menu_items_parquet_df = spark.read.format("parquet") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("fs.s3a.access.key", access_key) \
    .option("fs.s3a.secret.key", secret_key) \
    .load(s3_relative_path+object_menu_items_parquet)

raw_restaurant_csv_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("fs.s3a.access.key", access_key) \
    .option("fs.s3a.secret.key", secret_key) \
    .load(s3_relative_path+object_restaurant_csv)

raw_restaurant_parquet_df = spark.read.format("parquet") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("fs.s3a.access.key", access_key) \
    .option("fs.s3a.secret.key", secret_key) \
    .load(s3_relative_path+object_restaurant_parquet)


```

```python
#defining the DSFS volume
dbfs_relative_path = "/Volumes/dbacademy/dbuser_100426/data-volume-swiggy"
menu_items_csv_file = "/menu_items_csv"
menu_items_parquet_file = "/menu_items_parquet"
restaurant_csv_file = "/restaurant_csv"
restaurant_parquet_file = "/restaurant_parquet"
```

```python
#adding the current timestamp
from pyspark.sql.functions import current_timestamp

raw_menu_items_csv_df = raw_menu_items_csv_df.withColumn("ingestion_time", current_timestamp())
raw_menu_items_csv_df.write.format("delta").mode("append").save(dbfs_relative_path+menu_items_csv_file)

raw_restaurant_csv_df = raw_restaurant_csv_df.withColumn("ingestion_time", current_timestamp())
raw_restaurant_csv_df.write.format("delta").mode("append").save(dbfs_relative_path+restaurant_csv_file)

raw_menu_items_parquet_df = raw_menu_items_parquet_df.withColumn("ingestion_time", current_timestamp())
raw_menu_items_parquet_df.write.format("delta").mode("append").save(dbfs_relative_path+menu_items_parquet_file)

raw_restaurant_parquet_df = raw_restaurant_parquet_df.withColumn("ingestion_time", current_timestamp())
raw_restaurant_parquet_df.write.format("delta").mode("append").save(dbfs_relative_path+restaurant_parquet_file)
```

```python
# renaming the bronze finalized files
spark.read.format("delta").load(dbfs_relative_path+menu_items_csv_file)\
    .write.format("delta").mode("overwrite").saveAsTable("bronze_menu_items_csv_file")

spark.read.format("delta").load(dbfs_relative_path+menu_items_parquet_file)\
    .write.format("delta").mode("overwrite").saveAsTable("bronze_menu_items_parquet_file")

spark.read.format("delta").load(dbfs_relative_path+restaurant_csv_file)\
    .write.format("delta").mode("overwrite").saveAsTable("bronze_restaurant_csv_file")

spark.read.format("delta").load(dbfs_relative_path+restaurant_parquet_file)\
    .write.format("delta").mode("overwrite").saveAsTable("bronze_restaurant_parquet_file")
```
